```python
# ================================
# LangChain Demo (No API Required)
# ================================

"""
Note:
This notebook demonstrates core LangChain concepts such as prompt templates,
chains, memory, and tools.

LLM responses are simulated due to API key limitations.
"""

# ================================
# 1. Prompt Template
# ================================
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_template(
    "Explain {topic} in simple terms"
)

formatted_prompt = prompt.format(topic="LangChain")
print("Formatted Prompt:")
print(formatted_prompt)

print("\n" + "="*50 + "\n")


# ================================
# 2. Simulated LLM Function
# ================================
def fake_llm(input_text):
    return f"[Simulated LLM Response]: {input_text} is an important concept in AI systems."

response = fake_llm("LangChain")
print("LLM Output:")
print(response)

print("\n" + "="*50 + "\n")


# ================================
# 3. Simple Chain (Prompt → LLM)
# ================================
def simple_chain(topic):
    formatted = prompt.format(topic=topic)
    return fake_llm(formatted)

print("Chain Output:")
print(simple_chain("Machine Learning"))

print("\n" + "="*50 + "\n")


# ================================
# 4. Memory Simulation
# ================================
class SimpleMemory:
    def __init__(self):
        self.history = []

    def add(self, user_input, response):
        self.history.append((user_input, response))

    def show_history(self):
        return self.history


memory = SimpleMemory()

def chat(user_input):
    response = fake_llm(user_input)
    memory.add(user_input, response)
    return response

print("Conversation:")
print(chat("My name is Devyani"))
print(chat("What is my name?"))

print("\nChat History:")
print(memory.show_history())

print("\n" + "="*50 + "\n")


# ================================
# 5. Tool Example (Calculator)
# ================================
def calculator(expression):
    try:
        return eval(expression)
    except:
        return "Invalid expression"

print("Tool Output:")
print("20 * 5 =", calculator("20 * 5"))

print("\n" + "="*50 + "\n")


# ================================
# 6. Agent Simulation
# ================================
def simple_agent(query):
    if "calculate" in query.lower():
        expr = query.lower().replace("calculate", "").strip()
        return calculator(expr)
    else:
        return fake_llm(query)

print("Agent Output:")
print(simple_agent("calculate 10 + 15"))
print(simple_agent("Explain LangChain"))

print("\n" + "="*50 + "\n")


# ================================
# 7. Final Insight
# ================================
print("This notebook demonstrates how LangChain components work together:")
print("- Prompt Templates")
print("- Chains")
print("- Memory")
print("- Tools")
print("- Agents")
```
